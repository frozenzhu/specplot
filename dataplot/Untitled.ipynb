```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
from scipy import stats

def f_1(x, A, B, C):
    return A*np.exp(-x/B)+C
fig,ax=plt.subplots(figsize=(10,8))

x0 = [1, 2, 3, 4, 5]
y0 = [1, 3, 8, 18, 36]
ax.scatter(x0[:], y0[:], 25, "red")
A1,B1,C1 = optimize.curve_fit(f_1, x0, y0, p0=[1,-1,-1])[0]
print(A1, B1, C1)
y2=[]
y1=[]
x1 = np.arange(0, 6, 0.01)
for i in x0:
    y2.append(f_1(i,A1,B1,C1))
for j in x1:
    y1.append(f_1(j,A1,B1,C1))
#print(y2)
r=np.corrcoef(y2,y0)[0][1]**2
print(r)
x2,p=stats.chisquare(y0,f_exp=y2)
print(x2,p)
plt.plot(x1, y1, "blue")
print(np.searchsorted(y0,10))
col_labels = ['Chi-Square','p-Value','r-square']
table_vals = [[28],[29],[30]]
my_table = plt.table(cellText=table_vals, colWidths=[0.1]*3, rowLabels=col_labels,loc='best',
                    rowLoc='center', colLoc='center')
```
