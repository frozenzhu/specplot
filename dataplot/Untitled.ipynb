```python
import os
import numpy as np

def norm_by_pump(Data, spec_path, pump_wavelength):
    filepath = spec_path
    rspec = np.loadtxt(spec_path,delimiter='	')
    a = rspec[0]
    rspec1 = rspec[:,1]
    print('rspec1\n',rspec1)
    #rspec = np.loadtxt(filepath,delimiter=';',skiprows = 7, usecols = (0,1))
    rspec_norm = rspec1/np.max(np.absolute(rspec1))
#    np.savetxt(fname='./rspec_normed.txt',X=rspec_norm)
#    np.savetxt(fname='./pump_wavelength',X=pump_wavelength)
#    print(pump_wavelength[1])
#    print(np.shape(Data))
    data_norm = np.zeros((len(Data[:,0]),len(pump_wavelength)))
    print('pump_wavelength\n',pump_wavelength)
    for i in range(np.size(pump_wavelength)):
        wavelength_index = np.searchsorted(a, pump_wavelength[i])
        temp = Data[:,i]*2/(rspec_norm[wavelength_index]+rspec_norm[wavelength_index+1])
        data_norm[:,i]=temp
        #print((rspec_norm[wavelength_index]+rspec_norm[wavelength_index+1]))
    return data_norm

rdata = np.loadtxt(r'D:\课题组\OneDrive\课题组\OPV\2D\Vis\Y6\20220816-magic-680-40uw\220816_195407_0ps.2dv')
pump_wavelength = Delay = rdata[1:,0]
pump_min = 650
pump_max = 695
probe_min = 450
probe_max = 720
probe_wavelength = rdata[0,1:]
pump_wavelength = Delay = rdata[1:,0]
pump_min_index = np.searchsorted(pump_wavelength, pump_min)
pump_max_index = np.searchsorted(pump_wavelength, pump_max)
pump_wavelength = pump_wavelength[pump_min_index:pump_max_index]
probe_min_index = np.searchsorted(probe_wavelength, probe_min)
probe_max_index = np.searchsorted(probe_wavelength, probe_max)
probe_wavelength = probe_wavelength[probe_min_index:probe_max_index]
data = rdata [pump_min_index:pump_max_index,probe_min_index:probe_max_index]
spec_path = r'D:\课题组\OneDrive\课题组\OPV\2D\Vis\Y6\20220816-magic-680-40uw\PumpSpec\20220816-1833-Y-0nm.dat'

norm_by_pump(data, spec_path, pump_wavelength)


```
