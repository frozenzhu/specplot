```python
# read data from files

file = '../PT1-Raman/pt1-raman.CSV'

# with open(file, 'r', encoding='utf-8') as f:
#     sdata = f.read()

sdata = np.loadtxt(file, delimiter=',')
```

```python
# parse data

xs = sdata[:,0]
ys = sdata[:,1]

# effx = min([(len(i) - 1) if ('.' in i) else len(i) for i in xs])
# effy = min([(len(i) - 1) if ('.' in i) else len(i) for i in ys])
effx = 'na'
effy = 'na'


ptemp = '''Data parsed. 
Found {lenx} x values, {leny} y values.
Estimated effective figures: x: {effx}, y: {effy}
First 5 rows are:'''

print(ptemp.format(lenx=len(xs), leny=len(ys), effx=effx, effy=effy))
for i in range(5):
    print(xs[i], ys[i])


# print(s)
```

```python
# set plot details

fig, ax = plt.subplots()

ax.plot(xs, ys, linewidth=1, label='Solid PT1 on CF2', color='black')

ax.set(xlabel='Wavenumber (cm-1)', 
       ylabel='Absorption',
       title='Raman Spectrum')

ax.set(ylim=(0,40000),xlim=(800, 1600))

# ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(50)) 
# ax.xaxis.set_major_formatter(matplotlib.ticker.FormatStrFormatter('%1.1f')) 
# ax.yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(0.1)) 
# ax.yaxis.set_major_formatter(matplotlib.ticker.FormatStrFormatter('%1.2f')) 

# for FTIR, it is common to invert x axis (wavenumbers)
ax.invert_xaxis()

ax.grid()

plt.show()
```
