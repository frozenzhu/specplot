```python
# drawing utils
%matplotlib notebook
from matplotlib import pyplot as plt
import matplotlib.ticker

# math utils
import numpy as np

# statistics utils
import pandas as pd
```

```python
# read data from files

file = '../PT1-uvvis/7.txt'

with open(file, 'r', encoding='gb2312') as f:
    sdata = f.read()
```

```python
# parse data

xs = sdata.split()[5::2]
ys = sdata.split()[6::2]

effx = min([(len(i) - 1) if ('.' in i) else len(i) for i in xs])
effy = min([(len(i) - 1) if ('.' in i) else len(i) for i in ys])

xs = list(map(float, xs))
ys = list(map(float, ys))

xs = np.array(xs)
ys = np.array(ys)


ptemp = '''Data parsed. 
Found {lenx} x values, {leny} y values.
Estimated effective figures: x: {effx}, y: {effy}
First 5 rows are:'''

print(ptemp.format(lenx=len(xs), leny=len(ys), effx=effx, effy=effy))
for i in range(5):
    print(xs[i], ys[i])
```

```python
# set plot details

fig, ax = plt.subplots()

ax.plot(xs, ys, linewidth=1)

ax.set(xlabel='Wavelength (nm)', 
       ylabel='Absorption',
       title='UV-Vis Spectrum')

ax.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(50)) 
ax.xaxis.set_major_formatter(matplotlib.ticker.FormatStrFormatter('%1.2f')) 
ax.yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(0.1)) 
ax.yaxis.set_major_formatter(matplotlib.ticker.FormatStrFormatter('%1.2f')) 

ax.grid()

plt.show()
```
